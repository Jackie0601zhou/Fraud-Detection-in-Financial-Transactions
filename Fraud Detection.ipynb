```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
data.head()  # check front 5 line
data.info()  
data.describe()  
```

```python
data.isnull().sum()  # 检查缺失值
print("Check duplicates:", data.duplicated().sum())
```

```python
data = data.drop_duplicates() # delete duplicates
```

```python
# Since the Time and Amount features have not been scaled, they need to be standardized.

if 'Time' in data.columns and 'Amount' in data.columns:
    print("Columns 'Time' and 'Amount' found. Proceeding with standardization...")
    from sklearn.preprocessing import StandardScaler

    data['Scaled_Amount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1, 1))
    data['Scaled_Time'] = StandardScaler().fit_transform(data['Time'].values.reshape(-1, 1))

    # Delete "Time" and "Amount" Column
    data = data.drop(['Time', 'Amount'], axis=1)

    # reordered columns
    scaled_amount = data.pop('Scaled_Amount')
    scaled_time = data.pop('Scaled_Time')
    data.insert(0, 'Scaled_Amount', scaled_amount)
    data.insert(1, 'Scaled_Time', scaled_time)

    print("Standardization complete and columns reordered.")
else:
    print("Columns 'Time' and 'Amount' have already been processed.")
```

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Class', data=data)
plt.title('Class Distribution')
plt.show()

print(data['Class'].value_counts())
```

```python
data.hist(figsize=(20, 20))
plt.show()
```

```python
corr_matrix = data.corr()

plt.figure(figsize=(12, 9))
sns.heatmap(corr_matrix, vmax=0.8, square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Due to the severe imbalance in the dataset, processing is required to improve model performance.
# Under-sampling

# divide data into normal and fraud
fraud_data = data[data['Class'] == 1]
normal_data = data[data['Class'] == 0]


normal_data_sampled = normal_data.sample(n=len(fraud_data))

# Consolidation of data
balanced_data = pd.concat([fraud_data, normal_data_sampled])

# Disrupting data
balanced_data = balanced_data.sample(frac=1).reset_index(drop=True)
```

```python
# Over-sampling
from imblearn.over_sampling import SMOTE

X = data.drop('Class', axis=1)
y = data['Class']

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
```

```python
# feature selection
# Calculation of relevance
corr_matrix = data.corr()
top_features = corr_matrix['Class'].abs().sort_values(ascending=False).head(10)
print(top_features)
```

```python
# data segmentation：8:2
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)


print(f"Training set size: {X_train.shape}, Testing set size: {X_test.shape}")
```

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression


lr_model = LogisticRegression(max_iter=1000, random_state=42)


lr_model.fit(X_train, y_train)

# predication
y_pred_lr = lr_model.predict(X_test)
```

```python
# Random forest

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X_train, y_train)

# pred
y_pred_rf = rf_model.predict(X_test)
```

```python
# XGBoost
from xgboost import XGBClassifier

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

xgb_model.fit(X_train, y_train)

# pred
y_pred_xgb = xgb_model.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, accuracy_score

def evaluate_model(y_test, y_pred, model_name):
    print(f"=== {model_name} ===")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\n")
```

```python
evaluate_model(y_test, y_pred_lr, "Logistic Regression")
```

```python
evaluate_model(y_test, y_pred_rf, "Random Forest")
```

```python
evaluate_model(y_test, y_pred_xgb, "XGBoost")
```

```python
# Comparisons
from sklearn.metrics import f1_score, recall_score

#recall score
recall_lr = recall_score(y_test, y_pred_lr)
recall_rf = recall_score(y_test, y_pred_rf)
recall_xgb = recall_score(y_test, y_pred_xgb)

# F1 score
f1_lr = f1_score(y_test, y_pred_lr)
f1_rf = f1_score(y_test, y_pred_rf)
f1_xgb = f1_score(y_test, y_pred_xgb)


print(f"Logistic Regression - Recall: {recall_lr}, F1 Score: {f1_lr}")
print(f"Random Forest - Recall: {recall_rf}, F1 Score: {f1_rf}")
print(f"XGBoost - Recall: {recall_xgb}, F1 Score: {f1_xgb}")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(xgb_model, X_test, y_test)
plt.title('Confusion Matrix - XGBoost')
plt.show()
```

```python
y_scores = xgb_model.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
plt.plot(fpr, tpr, label='XGBoost')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
auc_value = roc_auc_score(y_test, y_scores)
print(f"AUC: {auc_value}")
```
